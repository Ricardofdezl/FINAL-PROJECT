```python
import pandas as pd
```

```python
df_women = pd.read_csv("MD_women_dataset.csv")
df_women["section"] = "woman"
df_women
```

```python
df_men = pd.read_csv("MD_men_dataset.csv")
df_men["section"] = "man"
df_men
```

```python
massimo_dutti_df = pd.concat([df_women, df_men], ignore_index=True)
massimo_dutti_df
```

```python
massimo_dutti_df = massimo_dutti_df.rename(columns={
    'links': 'url',
    'Price': 'price',
    'Title': 'name',
    'Category': 'department'})
massimo_dutti_df = massimo_dutti_df.dropna(subset=['name', 'price'])
massimo_dutti_df
```

```python
massimo_dutti_df['category'] = massimo_dutti_df['name'].str.split().str[0]

massimo_dutti_df
```

```python
massimo_dutti_df['price'] = massimo_dutti_df['price'].replace({'€': '', ',': '.'}, regex=True).astype(float)

massimo_dutti_df
```

```python
massimo_dutti_df['price'] = massimo_dutti_df['price'].replace({'€': '', ',': '.'}, regex=True).astype(float)
mean_price = massimo_dutti_df['price'].mean()

mean_price
```

```python
# Replace "man" with "Man" and "woman" with "Woman" in the 'section' column
massimo_dutti_df['section'] = massimo_dutti_df['section'].replace({'man': 'MAN', 'woman': 'WOMAN'})

# Display the first few rows to confirm the ch
massimo_dutti_df
```
